```python
#Fill all missing values using iterative imputer
imputer = IterativeImputer(KNeighborsRegressor())
imputed_data = imputer.fit_transform(num_data)
imputed_data = pd.DataFrame(imputed_data, columns = num_data.columns)
```

```python
Y = full_data['SalePrice'][full_data['train'] == 1].copy()
```

```python
df_train = full_data[full_data['train'] == 1]
df_train = df_train.drop(['train'],axis=1)
df_train.drop(['SalePrice'], axis = 1, inplace = True)

df_test = full_data[full_data['train'] == 0]
df_test = df_test.drop(['train'],axis=1)
df_test.drop(['SalePrice'], axis = 1, inplace = True)
```

```python
#Split the data into training and validations sets
X_train, X_val, Y_train, Y_val = train_test_split(df_train, Y, test_size = 0.30, random_state = 4)
```

```python
print('Length Of X_train and X_val are {} and {}'.format(len(X_train), len(X_val)))
```

```python
#Choose best parameters for XGBRegressor using Gridsearchcv
param_grid = {
    'max_depth' : [3 ,4 ,5],
    'learning_rate' : [0.01, 0.1, 0.5, 1],
    'gamma' : [0, 0.25, 0.5],
    'reg_alpha' : [0.4, 0.6, 0.8],
    'reg_lambda' : [0.4, 0.6, 0.8],
    'scala_pos_weight' : [1, 2, 3]
}

optimal_params = GridSearchCV(
        estimator = xgb.XGBRegressor(objective='reg:linear',
                                     colsample_bylevel=1,
                                     subsample=0.8,
                                     colsample_bytree=0.6
                                    ),
        param_grid = param_grid,
        scoring = 'neg_mean_squared_error',
        verbose = 2,
        cv = 3,
        n_jobs = -1)
```

```python
grid_result = optimal_params.fit(X_train, Y_train)
best_params = grid_result.best_params_
```

```python
best_params
```

```python
xgb_reg = xgb.XGBRegressor( booster='gbtree', colsample_bylevel=1,
             colsample_bynode=1, colsample_bytree=0.6, gamma=0,
             importance_type='gain', learning_rate=0.1, max_delta_step=0,
             max_depth=3, min_child_weight=1.5, n_estimators=2400,
             n_jobs=1, nthread=None, objective='reg:linear',
             reg_alpha=0.8, reg_lambda=0.4, scale_pos_weight=1, 
             silent=None, subsample=0.8, verbosity=1)
```

```python
#Fitting
xgb_reg.fit(X_train, Y_train, verbose = 1, early_stopping_rounds = 30, eval_metric = 'rmse', eval_set = [(X_val, Y_val)])
```
